```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import subprocess
```

```python
notepad = subprocess.Popen(['./example', 'input.q10', 'output.txt'])
```

```python
class FileReader(object):
    
    def __init__(self):
        self.direct = 0
        self.matr = 0
        
    def get_matrix(self, direct):
        self.direct = direct
        f = open(self.direct, 'r')
        lines = f.readlines()
        size = (len(lines), len(lines[1].split(" "))-1)
        matr = np.zeros(size)
        for i in range(len(lines)):
            matr[i] = np.array((lines[i].split(" "))[:-1])
        self.matr = matr
        f.close()
        
    def show_matr(self):
        plt.figure(figsize=(10,10))
        plt.imshow(self.matr)
        plt.show()
        
```

```python
FR = FileReader()
```

```python
FR.get_matrix('output.txt')
FR.show_matr()
```
